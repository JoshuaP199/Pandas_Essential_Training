```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008 using matplotlib and seaborn. How can you use colormaps to give the data more meaning?

```python
data = pd.read_csv("Olympics.csv", skiprows = 4)
data.head()
```

```python
cm = data[(data.Edition == 2008) & (data.NOC == "CHN")]
cm
```

```python
cm.Gender.value_counts()
```

```python
cm.Medal.value_counts()
```

```python
cm.Gender.value_counts().plot(kind = 'pie');
```

```python
sns.countplot(x = 'Gender', data = cm, hue = 'Medal');
```

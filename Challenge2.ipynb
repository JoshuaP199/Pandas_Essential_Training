### In which events did Jesse Owens win a medal?

```python
data = pd.read_csv("olympics.csv", skiprows = 4)
```

```python
jess = data[data.Athlete.str.contains('OWENS')]
jess
#he won 4
#CORRECT
```

```python
jess.Event.value_counts()
```

### Which country has won the most men's gold medals in singles badminton over the years? Sort the results alphabetically by the player's  names.

```python
bmgs = data[(data.Sport == 'Badminton') & (data.Gender == 'Men')
     & (data.Medal == 'Gold') & (data.Event == 'singles')]
bmgs.sort_values(by = 'NOC')
#CORRECT
#CHN and INA have the most 
```

```python
bmgs.sort_values(by = 'Athlete')
#CORRECT
```

### Which three countries have won the most medals in recent years (from 1984 to 2008)?

```python
year = data[(data.Edition >= 1984) & (data.Edition < 2008)]
year
```

```python
year.NOC.value_counts().head(3)
```

### Display the male gold medal winners for the 100m Track & Field sprint event over the years. List the results starting with the most recent. Show the Olympic City, Edition, Athlete and the country they represent.

```python
male = data[(data.Gender == 'Men') & (data.Medal == 'Gold') & (data.Event == '100m')]
male
```

```python
male.sort_values("Edition", ascending = False)[['City', 'Edition', 'Athlete', 'NOC']]
```
